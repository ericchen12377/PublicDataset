```python
import pandas as pd
```

```python
import os 
work_path = os.getcwd() 
raw_path = work_path + "\\raw"
processed_path = work_path + "\\processed"
```

```python
df_train = pd.read_csv(
    raw_path+"\\albert\\albert_train.data", header=None, sep=' ')
df_train.drop([78], axis=1, inplace=True)
df_train.head()
```

```python
df_train.shape
```

```python
df_tag = pd.read_csv(raw_path+"\\albert\\albert_train.solution", names=['TAG'],
                     header=None, sep=' ')
df_tag
```

```python
df_train = pd.concat([df_train, df_tag], axis=1)
df_train.shape
```

```python
df_train.to_csv(processed_path+"\\albert.csv", header=True, index=None)
```
