```python
import pandas as pd
```

```python
import os 
work_path = os.getcwd() 
raw_path = work_path + "\\raw"
processed_path = work_path + "\\processed"
```

```python
df_train = pd.read_csv(raw_path+"\\arcene\\ARCENE\\arcene_train.data", header=None, sep=" ")
df_train.head()
```

```python
df_train.shape
```

```python
df_train_label = pd.read_csv(raw_path+"\\arcene\\ARCENE\\arcene_train.labels",
                             header=None, sep=" ", names=['TAG'])
df_train_label
```

```python
pd.concat([df_train,df_train_label], axis=1)
```

```python
df_valid = pd.read_csv(raw_path+"\\arcene\\ARCENE\\arcene_valid.data",
                 header=None, sep=" ")
df_valid.head()
```

```python
df_valid.shape
```

```python
df_valid_label = pd.read_csv(raw_path+"\\arcene\\ARCENE\\arcene_valid.labels",
                             header=None, sep=" ", names=['TAG'])
df_valid_label
```

```python
pd.concat([df_valid, df_valid_label], axis=1)
```

```python
pd.concat([pd.concat([df_train, df_train_label], axis=1), pd.concat([df_valid, df_valid_label], axis=1)], axis=0).reset_index(drop=True).to_csv(
    processed_path+"\\arcena.csv", header=True, index=None)
```
