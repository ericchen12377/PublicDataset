```python
import pandas as pd
```

```python
import os 
work_path = os.getcwd() 
raw_path = work_path + "\\raw"
processed_path = work_path + "\\processed"
```

```python
df_train = pd.read_csv(raw_path+"\\volkert\\volkert_train.data", header=None, sep=' ')
df_train.drop([180], axis=1, inplace=True)
df_train.head()
```

```python
df_tag = pd.read_csv(
    raw_path+"\\volkert\\volkert_train.solution", header=None, sep=' ', names=['class-1','class-2','class-3','class-4','class-5','class-6','class-7','class-8','class-9','class-10', 'space'])
df_tag.drop(['space'], axis=1, inplace=True)
df_tag
```

```python
df_train.shape, df_tag.shape
```

```python
df_train = pd.concat([df_train, df_tag], axis=1)
df_train
```

```python
df_train.to_csv(processed_path+"\\volkert.csv", header=True, index=None)
```
